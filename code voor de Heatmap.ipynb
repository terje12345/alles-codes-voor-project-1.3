```python
# deze cel maakt de heatmap.

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.colors import LinearSegmentedColormap

# De gemiddelde nagalmtijden
nagalmtijden = [
2.3, 1.0, 1.4, 0.8, 1.3, 2.2, 1.3, 2.3, 1.6, 2.5, 3.1, 1.6, 1.9, 1.7, 1.5, 
    2.6, 2.1, 1.1, 1.4, 1.4, 2.0, 1.6, 2.1, 1.1, 2.3, 1.4,
    1.5, 1.2, 1.3, 0.6, 1.7, 1.4, 1.9, 1.3, 1.5, 1.4, 1.8, 0.9, 1.6, 1.0,
    1.4, 2.7, 1.2, 2.2, 2.1, 0.8, 1.2, 1.4, 1.3, 1.4, 1.7, 0.6, 0.6, 1.1, 1.6, 1.0, 1.6, 1.2, None, 1.0
]

# Data omzetten naar een 2D-array (10 rijen, 6 kolommen, laatste rij 5 kolommen)
heatmap_data = np.full((10, 6), np.nan)  # Maak een lege array met NaN-waarden
for i, value in enumerate(nagalmtijden):
    row = i // 6
    col = i % 6
    heatmap_data[row, col] = value

# Aangeven van de colormap (groen  geel bij 1  oranje  rood) 1 bij geel want 1 is net te hoog 
custom_cmap = LinearSegmentedColormap.from_list(
    "custom_cmap", 
    ["green", "yellow", "orange", "red"]
)

# het plotten van de heatmap 
plt.figure(figsize=(8, 6))
sns.heatmap(
    heatmap_data, 
    annot=True, 
    cmap=custom_cmap,  
    linewidths=0.5, 
    cbar=True, 
    vmin=np.nanmin(heatmap_data),  
    vmax=np.nanmax(heatmap_data)
)
# informatie aan de heatmap geven.
plt.title("Nagalmtijd Heatmap (Groen  Geel bij 1  Rood)")
plt.xlabel("Kolommen")
plt.ylabel("Rijen")
plt.show()
```
